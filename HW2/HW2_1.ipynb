```python
# 먼저 데이터 자르기
data1 = df.head(5)
data2 = data1[['pregnant', 'plasma']].values

print(data2)
print(data2.shape)
print(type(data2))

# 벡터 -> 배열로 변경 (굳이 안바꿔도 됨. 벡터, 배열 다 append 가능)
# data3 = data2.tolist()

x1 = []
y1 = []

for i in range(data2.shape[0]) :
  x1.append(data2[i][0])
print(x1)

for i in range(data2.shape[0]) :
  y1.append(data2[i][1])
print(y1)

# 그래프 출력
plt.plot(x1, y1, "o")
plt.plot(x1, y1)
plt.show()
```
